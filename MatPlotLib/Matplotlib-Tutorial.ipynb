```python
df = pd.DataFrame({'Girl':['IU','Suzy','Punch','Kim So Hyun','Park Min Young','Rose',], 'Beauty':[10,9,9,10,9,8]})
```

```python
df
```

```python
plt.rc('font',size=20)

fig = plt.figure(figsize=(20,20))
axis = fig.add_subplot(1,1,1)

axis.set_xticklabels( df.Girl,rotation=90)
axis.bar(height = df.Beauty,x = df.Girl)
```

```python
plt.rc('font',size=20)

fig = plt.figure(figsize=(20,20))
axis = fig.add_subplot(1,1,1)

axis.barh(width = df.Beauty,y = df.Girl)
```

```python
movies = pd.read_csv('Movie-Rating.csv')
```

```python
movies.head()
```

```python

movies.columns = ['Film','Genre','CriticRating','AudienceRating','BudgetMillions','Year']
plt.rc('figure',figsize=(30,30))
plt.yticks(list(range(0,150,20)))
plt.hist([movies[movies.Genre == 'Action'].BudgetMillions, \
          movies[movies.Genre == 'Drama'].BudgetMillions, \
          movies[movies.Genre == 'Thriller'].BudgetMillions, \
          movies[movies.Genre == 'Comedy'].BudgetMillions], bins = 20,\
          stacked = True)

plt.legend(['Action','Drama','Thriller','Comedy'])
```

# Box Plot


```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

plt.title('CriticRating : Action v/s Drama')
plt.ylabel('CriticRating')

plt.boxplot([movies.loc[movies.Genre == 'Action']['CriticRating'],\
             movies[movies.Genre == 'Drama']['CriticRating']],\
             labels = ['Action', 'Drama'],\
             patch_artist = True,\
             medianprops={'linewidth': 2},\
             boxprops = {'facecolor':'yellow'})
```

```python
time = list(range(0,30))
speed = 100 #metres/sec
acceleration = -5 #meters/sec
distance = list()
stoptime = 0
for t in time:
    
    s = speed*t + 0.5*(acceleration)*(t**2)
    speed = speed + acceleration*t
    if(s < 0):
        print(f"car stopped at time = {t-1}")
        stoptime = t
        break
    distance.append(s)
    
distance
    
```

```python
plt.fill_between(range(stoptime), distance, color = 'pink' )
```
