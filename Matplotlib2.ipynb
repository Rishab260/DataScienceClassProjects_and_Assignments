```python
def to_intW(x):
    if(type(x) == str):
        return int(x.strip('lbs'))
    else:
        return 0
```

```python
fifa['Weight'] = fifa['Weight'].apply(to_intW)
```

```python
fifa['Weight']
```

```python
fifa['Weight'].describe()
```

```python
Filter_CountZeros = fifa.Weight == 0
NumberOfZeros =  fifa[Filter_CountZeros] #must be subtracted

Filter_Over = (fifa['Weight'] > 200)
Filter_MediumHigh = (fifa.Weight <= 200) & (fifa.Weight > 150)
Filter_MediumLow = (fifa.Weight <= 150) & (fifa.Weight > 110)
Filter_Under = (fifa.Weight <= 100 ) & (fifa.Weight != 0)
```

```python
print(len(fifa[Filter_Over]),len(fifa[Filter_MediumHigh]) , len(fifa[Filter_MediumLow]), len(fifa[Filter_Under]))
```

```python
plt.style.use('ggplot')
plt.rcParams['font.size'] = 30
plt.pie( [len(fifa[Filter_Over]), len(fifa[Filter_MediumHigh]) , len(fifa[Filter_MediumLow])],
        labels=['OverWeight','MediumHigh','MediumLow'])
```
